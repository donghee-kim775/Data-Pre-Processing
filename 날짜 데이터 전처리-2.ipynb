# dataframe['해당컬럼'].apply(함수)
#### .apply(함수) : 행 또는 열 또는 전체의 셀에 원하는 연산을 지원하는 함수

```python
def date_function(x):
   return x[0:10]

df2['배송시작일(datetime)']=pd.to_datetime(df2['배송시작일'].apply(date_function))
df2['배송완료일(datetime)']=pd.to_datetime(df2['배송완료일'].apply(date_function))
df2
```

#### -> 앞에 9글자만 가져오면 날짜 형태로 바꿀 수 있음

```python
df2['배송시간']=df2['배송완료일(datetime)']-df2['배송시작일(datetime)']
df2.head()
```

```python
def date_function2(x): # 주문일은 정수형이라 인덱싱이 안됨
    result=str(x) # 문자형으로 바꿔주고 인덱싱해주면 됨
    return result[0:4]+'-'+result[4:6]+'-'+result[6:8]

pd.to_datetime(df2['주문일'].apply(date_function2))
```
