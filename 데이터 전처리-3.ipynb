## 내가 궁금했던점 : loc 와 iloc의 차이?
#### loc : 데이터프레임의 행이나 컬럼에 label이나 boolean array로 접근
#### iloc : 데이터프레임의 행이나 컬럼에 인덱스값으로 접근

```python
cond1=(df1['연령대']=='40대')
cond2=(df1['성별']=='남')
df1.loc[cond1 & cond2]
```

# 연습

```python
# 30대 여성들의 데이터만 추출해서 구매금액이 높은순서대로 파일만들기
# 1번째 30대 여성 데이터 추출
# 2번째 구매금액 높은순서로 정렬
# 파일로 저장
```

```python
cond1=(df1['연령대']=='30대이하')
cond2=(df1['성별']=='여')

df3=df1[cond1 & cond2].sort_values(by='구매금액',ascending=False)
df3.shape
```

```python
# csv데이터 저장-> .to_csv()
df3.to_csv('result.csv', encoding='cp949')    # encoding='cp949'->한글 인코딩 지원하는 인코더 입력하기->글자가 깨지지않음
```
